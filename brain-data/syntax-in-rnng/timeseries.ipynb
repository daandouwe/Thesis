```python
X = []
with open('alice_data_shared/6mm/s18-timecourses.txt') as fin:
    for line in fin:
        x = [float(x) for x in line.split()]
        X.append(x)
X = np.array(X).T
```

```python
plt.plot(X[0])
plt.show()
```

```python
for i in range(X.shape[0]):
    plt.plot(X[i])
plt.show()
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts) / box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

n = 50
for i in range(X.shape[0]):
    X_ = smooth(X[i], n)
    plt.plot(X_[n:-n])
    plt.show()
```
