```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 100)
u = np.sin(x)
v = np.cos(x)
```

```python
plt.plot(x, u)
plt.plot(x, v)
```

```python
v_ = v - (v @ u) / (u @ u) * u
v_ @ u
```

```python
plt.plot(x, u)
plt.plot(x, v)
plt.plot(x, v_)
```
