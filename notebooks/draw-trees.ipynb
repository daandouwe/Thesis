## Negative polarity

### CRF parser

```python
npi_pos = "(S (NP (NP The talk) (SBAR (S (NP I) (VP gave)))) (VP did not (VP appeal (PP to (NP anybody)))) .)"
npi_neg = "(S (NP The talk) (NP I) (VP did not (VP give appealed (PP to (NP anybody)))) .)"
```

```python
Tree.fromstring(npi_pos)
```

```python
Tree.fromstring(npi_neg)
```

```python
npi_pos = "(S (NP (NP The talk) (SBAR (S (NP I) (VP gave)))) (VP did not (VP appeal (PP to (NP anybody)))) .)"
npi_neg = "(S (NP (NP The talk) (SBAR (S (NP I) (VP did not (VP give))))) (VP appealed (PP to (NP anybody))) .)"
```

## Garden path sentences

The sentence 

    The student forgot the solution was in the back of the book.

was taken from Jurafsky and Martin chapter 12 of 3rd edition.

The correct parse is found by the generative rnng, in some occasions!

```python
crf_tree = "(S (NP The student) (VP forgot (SBAR (S (NP the solution) (VP was (PP in (NP (NP the back) (PP of (NP the book)))))))) .)"
disc_tree = "(S (NP (NP The student forgot) (SBAR (S (NP the solution) (VP was)))) (PP in (NP (NP the back) (PP of (NP the book)))) .)"
gen_tree = "(S (NP The student) (VP forgot (SBAR (S (NP the solution) (VP was (PP in (NP (NP the back) (PP of (NP the book)))))))) .)"
```

```python
Tree.fromstring(crf_tree)
```

```python
Tree.fromstring(disc_tree)
```

```python
Tree.fromstring(gen_tree)
```

Another one from Hale (2001):

    The banker told about the buy-back resigned .


```python
crf_tree = "(S (NP (NP The banker) (VP told (PP about (NP the buy-back)))) (VP resigned) .)"
disc_tree = "(S (NP The banker) (VP told (SBAR about (S (NP the buy-back) (VP resigned)))) .)"
gen_tree = "(S (NP The banker) (VP told (PP about (NP the buy-back resigned))) .)"
```

```python
Tree.fromstring(crf_tree)
```

```python
Tree.fromstring(disc_tree)
```

```python
Tree.fromstring(gen_tree)
```
