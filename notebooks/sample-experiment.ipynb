```python
import os

import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
FOLDER = '../out/sample-experiment'

dataframes = []
for folder in os.listdir(FOLDER):
    if folder.endswith('.tsv'):
        dataframes.append(
            pd.read_csv(os.path.join(FOLDER, folder), sep='\t'))
```

```python
results = pd.concat(dataframes)
```

```python
sns.violinplot(
    data=results, x='num-samples', y='perplexity', hue='temp', palette='Set2', inner='stick', split=True)
```

```python
sns.boxplot(data=results, x='num-samples', y='perplexity')
```
