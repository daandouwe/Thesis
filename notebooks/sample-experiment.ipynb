```python
import os

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
FOLDER = '../out/sample-experiment'

disc_dfs, crf_dfs = [], []
for folder in os.listdir(FOLDER):
    if folder.endswith('.tsv'):
        if'disc' in folder:
            disc_dfs.append(pd.read_csv(os.path.join(FOLDER, folder), sep='\t'))
        else:
            crf_dfs.append(pd.read_csv(os.path.join(FOLDER, folder), sep='\t'))
            
```

```python
ALL = [1, 10, 25, 50, 75, 100]
SOME = [10, 25, 50, 75, 100]

PP_DIR = os.path.join('figures', 'perplexity')
F_DIR = os.path.join('figures', 'fscore')

os.makedirs(PP_DIR, exist_ok=True)
os.makedirs(F_DIR, exist_ok=True)
```

Mark difference between discriminative models (not properly done in script).

```python
disc = pd.concat(disc_dfs)
crf = pd.concat(crf_dfs)

disc['model'] = 'disc-rnng'
crf['model'] = 'crf'

both = pd.concat([crf, disc])

disc10 = disc[disc['temp'] == 1.0]
disc08 = disc[disc['temp'] == 0.8]

crf_disc10 = pd.concat([crf, disc10])
crf_disc08 = pd.concat([crf, disc08])
```

## Perplexity

Disc temperature

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.violinplot(
    data=disc, x='num-samples', y='perplexity', hue='temp', palette='Set2', inner='stick', order=SOME)
plt.savefig(os.path.join(PP_DIR, 'disc_temp.pdf'))

fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(
    data=disc, x='num-samples', y='perplexity', hue='temp', palette='Set2', order=SOME)
plt.savefig(os.path.join(PP_DIR, 'box_disc_temp.pdf'))

fig, ax = plt.subplots(figsize=(10, 5))
sns.violinplot(
    data=disc, x='num-samples', y='perplexity', hue='temp', palette='Set2', inner='stick', order=ALL)
plt.savefig(os.path.join(PP_DIR, 'all_disc_temp.pdf'))

fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(
    data=disc, x='num-samples', y='perplexity', hue='temp', palette='Set2', order=ALL)
plt.savefig(os.path.join(PP_DIR, 'box_all_disc_temp.pdf'))
```
